## Whisper 사용하기

```python
import whisper

# 모델 로드
# tiny < base < small < medium < large < turbo
base_model = whisper.load_model("base")
small_model = whisper.load_model("small")
```

```python
audio_path = "./resources/회의.mp3"  # 같은 폴더에 있다고 가정

import time 
import warnings
warnings.filterwarnings('ignore')

print("Base Model 사용")
start_time = time.time()
base_result = base_model.transcribe(audio_path)
end_time = time.time()

elasped_time = end_time - start_time
print(f"걸린 시간: {elasped_time}")

print("="*50)

print("Small Model 사용")
start_time = time.time()
small_result = small_model.transcribe(audio_path)
end_time = time.time()

elasped_time = end_time - start_time
print(f"걸린 시간: {elasped_time}")
```

```python
for seg in base_result["segments"]:
    print(seg["text"])
```

```python
for seg in small_result["segments"]:
    print(seg["text"])
```

```python
import json

# Whisper 결과에서 segments 추출
segments = small_result["segments"]

# segment별 저장 (json)
full_script = ""
segments_json = []
for seg in segments:
    segment_data = {"start": seg["start"], "end": seg["end"], "text": seg["text"]}
    # print(segment_data)
    segments_json.append(segment_data)

    full_script += seg["text"] + "\n"

with open("./output/whisper_output_segments.json", "w", encoding="utf-8") as f:
    json.dump(segments_json, f, ensure_ascii=False, indent=2)

with open("./output/whisper_full_script.txt", "w", encoding="utf-8") as f:
    f.write(full_script)
    # print(full_script)
```
